# What Are Outliers?

Outliers are **data points that significantly differ from other observations in a dataset**. They are unusually high or low values compared to the majority of data and can arise due to:

* Natural variation  
* Measurement errors  
* Data entry mistakes  



## Example

Imagine you collected math test scores:


65, 70, 68, 72, 69, 95, 67, 400


Here, **400** is clearly an outlier—it doesn’t align with typical test scores.



## Why Outliers Matter

* **Skew Analysis:** Outliers can distort statistical analyses like mean and standard deviation.  
* **Impact Models:** In machine learning, outliers can reduce model accuracy, especially for algorithms sensitive to data distribution (like linear regression).  
* **Real Insights:** Sometimes outliers reflect rare but important events (like detecting fraud).  



## How to Handle Them

1. **Detect:** Using boxplots, scatterplots, Z-score, or IQR (Interquartile Range) methods.  
2. **Investigate:** Decide if it’s a mistake or genuine data.  
3. **Treat:**

   * Remove (if error or irrelevant)  
   * Cap or transform (if extreme but real)
